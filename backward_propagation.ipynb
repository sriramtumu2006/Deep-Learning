# Diabetes Risk Prediction (Binary Classification)

```python
import numpy as np 
np.random.seed(0)
x=np.array([
    [120,85,150,95], #Glucose
    [25,30,35,22] #BMI
    ])

#Normalize the data
x=x/200

#Output : 1=High risk, 0=Low risk
y=np.array([[1,0,1,0]])
print(" x shape:",x.shape)
print(" y shape:",y.shape)
```

# Sigmoid function

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

# Initialize the parameters

```python
# Hidden Layer(3 neurons,2 inputs)
w1=np.random.randn(3,2)*0.1
b1=np.zeros((3,1))

# Output Layer(1 neuron,3 hidden inputs)
w2=np.random.randn(1,3)*0.1
b2=np.zeros((1,1))
```

# Forward Propagation

```python
z1=np.dot(w1,x)+b1
a1=sigmoid(z1)
z2=np.dot(w2,a1)+b2
a2=sigmoid(z2)
print("Predicted output:",a2)
```

# Loss Function

```python
m=y.shape[1] #number of examples
loss=-1/m*np.sum(y*np.log(a2)+(1-y)*np.log(1-a2))
print("Loss:",loss)
```

# Backward Propagation

```python
dz2=a2-y
dw2=np.dot(dz2,a1.T)/m
db2=np.sum(dz2,axis=1,keepdims=True)/m
da1=np.dot(w2.T,dz2)
dz1=da1*a1*(1-a1)
dw1=np.dot(dz1,x.T)/m
db1=np.sum(dz1,axis=1,keepdims=True)/m
print("dw2:",dw2)
print("db2:",db2)
print("dw1:",dw1)
print("db1:",db1)
```

# Update Parameters

```python
learning_rate=0.1
w2=w2-learning_rate*dw2
b2=b2-learning_rate*db2
w1=w1-learning_rate*dw1
b1=b1-learning_rate*db1
epoch=5000
for i in range(epoch):
    z1=np.dot(w1,x)+b1
    a1=sigmoid(z1)
    z2=np.dot(w2,a1)+b2
    a2=sigmoid(z2)
    m=y.shape[1]
    loss=-1/m*np.sum(y*np.log(a2)+(1-y)*np.log(1-a2))
    dz2=a2-y
    dw2=np.dot(dz2,a1.T)/m
    db2=np.sum(dz2,axis=1,keepdims=True)/m
    da1=np.dot(w2.T,dz2)
    dz1=da1*a1*(1-a1)
    dw1=np.dot(dz1,x.T)/m
    db1=np.sum(dz1,axis=1,keepdims=True)/m
    w2=w2-learning_rate*dw2
    b2=b2-learning_rate*db2
    w1=w1-learning_rate*dw1
    b1=b1-learning_rate*db1
    if i%1000==0:
        print(f"Epoch {i}, Loss: {loss:.4f}")
```
