# Perceptron(Binary Classification)

```python
import numpy as np
```

# Step 1 : Prepare Training Data

```python
x=np.array([1,2,3,4])
y=np.array([0,0,1,1])
x,y
```

# Step 2 : Initialize Parameters

```python
w=0.0
b=0.0
alpha=1 #learning rate
epochs=5
print(f"Initial weight: {w}, bias: {b}")   
```

# Step 3 : Define Activation Function(Step Function)

```python
def step_function(z):
    return 1 if z>=0 else 0
```

# Training Loop

```python
for epoch in range(epochs):
    print(f"Epoch : {epoch+1}")
    for i in range(len(x)):
        z=w*x[i]+b
        y_predicted=step_function(z)
        update=alpha*(y[i]-y_predicted)
        w+=update*x[i]
        b+=update
        error = y[i] - y_predicted
        print(f"Sample: {i} |   Target: {y[i]}  |   Predicted: {y_predicted}    |   Actual: {y[i] == y_predicted}   |   Error: {error}  |   Updated weight: {w} |   Updated bias: {b}")
```

# Final Model Parameters

```python
print(f"Final weight: {w}, bias: {b}")
```

# Testing Model

```python
test_values=3.1
z=w*test_values+b
y_predicted=step_function(z)
print(f"Test value: {test_values} | Predicted class: {y_predicted}")
```

# Plotting

```python
import matplotlib.pyplot as plt
plt.scatter(x, y, color='blue', label='Data Points')
boundary = -b / w
plt.axvline(x=boundary, color='red', label='Decision Boundary')
plt.xlabel('Feature (x)')
plt.ylabel('Target (y)')
plt.title('Perceptron Decision Boundary (1D)')
plt.legend()
plt.grid()
plt.show()
```

# Perceptron And Gate 

```python
x=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,0,0,1])
x,y
```

```python
w=np.zeros(2)
b=0
alpha=0.1
epochs=10
print(f"Initial weights: {w}, bias: {b}")
```

```python
for epoch in range(epochs):
    print(f"Epoch {epoch+1}")
    for i in range(len(x)):
        z=np.dot(w, x[i]) + b
        y_predicted=step_function(z)
        update=alpha*(y[i]-y_predicted)
        w+=update*x[i]
        b+=update
        error = y[i] - y_predicted
        print(f"Sample: {i} | Target: {y[i]} | Predicted: {y_predicted} | Actual: {y[i] == y_predicted} | Error: {error} | Updated weights: {w} | Updated bias: {b}")
```

```python
print(f"Final weight: {w}, bias: {b}")
test_values=np.array([[0,0],[0,1],[1,0],[1,1]])
for test in test_values:
    z=np.dot(w, test) + b
    y_predicted=step_function(z)
    print(f"Test value: {test} | Predicted class: {y_predicted}")
```

```python
plt.scatter(x[:,0], x[:,1], c=y, cmap='bwr', edgecolors='k', label='Data Points')
x1_values = np.linspace(0.0, 1.0, 100)
x2_values = -(w[0] * x1_values + b) / w[1]
plt.plot(x1_values, x2_values, color='red', label='Decision Boundary')
plt.xlabel('Feature 1 (x1)')
plt.ylabel('Feature 2 (x2)')
plt.title('Perceptron Decision Boundary for AND Logic Gate')
plt.legend()
plt.grid()
plt.show()
```

# Perceptron Or Gate

```python
x=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,1,1,1])
w=np.zeros(2)
b=0
alpha=0.1
epochs=10
print(f"Initial weights: {w}, bias: {b}")
for epoch in range(epochs):
    print(f"Epoch {epoch+1}")
    for i in range(len(x)):
        z=np.dot(w, x[i]) + b
        y_predicted=step_function(z)
        update=alpha*(y[i]-y_predicted)
        w+=update*x[i]
        b+=update
        error = y[i] - y_predicted
        print(f"Sample: {i} | Target: {y[i]} | Predicted: {y_predicted} | Actual: {y[i] == y_predicted} | Error: {error} | Updated weights: {w} | Updated bias: {b}")
print(f"Final weight: {w}, bias: {b}")
```

```python
print(f"Final weight: {w}, bias: {b}")
test_values=np.array([[0,0],[0,1],[1,0],[1,1]])
for test in test_values:
    z=np.dot(w, test) + b
    y_predicted=step_function(z)
    print(f"Test value: {test} | Predicted class: {y_predicted}")
```

```python
plt.scatter(x[:,0], x[:,1], c=y, cmap='bwr', edgecolors='k', label='Data Points')
x1_values = np.linspace(0.0, 1.0, 100)
x2_values = -(w[0] * x1_values + b) / w[1]
plt.plot(x1_values, x2_values, color='red', label='Decision Boundary')
plt.xlabel('Feature 1 (x1)')
plt.ylabel('Feature 2 (x2)')
plt.title('Perceptron Decision Boundary for OR Logic Gate')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
X = np.array([
    [1, 1],   # E1
    [1, 0],   # E2
    [0, 1],   # E3
    [0, 0]    # E4
])
y = np.array([1, 1, 1, 0])
weights = np.zeros(2)
bias = 0
learning_rate = 0.1
epochs = 20
for epoch in range(epochs):
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        prediction = 1 if linear_output >= 0 else 0 
        error = y[i] - prediction
        weights += learning_rate * error * X[i]
        bias += learning_rate * error
        print(f"Epoch: {epoch+1}, Sample: {i}, Target: {y[i]}, Predicted: {prediction}, Error: {error}, Updated Weights: {weights}, Updated Bias: {bias}")

print("Final Weights:", weights)
print("Final Bias:", bias)
x1_vals = np.linspace(-0.5, 1.5, 100)
if weights[1] != 0:
    x2_vals = (-weights[0] * x1_vals - bias) / weights[1]
else:
    x2_vals = np.zeros_like(x1_vals)
plt.scatter(X[y==1][:,0], X[y==1][:,1])
plt.scatter(X[y==0][:,0], X[y==0][:,1])
plt.plot(x1_vals, x2_vals)
plt.xlabel("Suspicious Keywords (x1)")
plt.ylabel("Unknown Sender (x2)")
plt.title("Perceptron Decision Boundary (Spam Detection)")
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.show()
```
